```julia
using JuMP
using CPLEX
```

```julia
m = 10
```

```julia
n = 30
```

```julia
p = 20
```

```julia
Q = rand(10:50, m, n)
```

```julia
C = rand(60:100, p, n)
```

```julia
model = Model(CPLEX.Optimizer)
```

```julia
@variable(model, x[i=1:m, j =1:p], lower_bound = 0, Bin)
```

```julia
@constraint(model, sum(x[:, j] for j in 1:p) .== 1)
```

```julia
@constraint(model, sum(x[i, :] for i in 1:m) .<= 1)
```

```julia
cost = sum(x[i, k] * C[k, j] * Q[i, j] for i in 1:m for j in 1:n for k in 1:p)
```

```julia
@objective(model, Min, cost)
```

```julia
optimize!(model)
```
